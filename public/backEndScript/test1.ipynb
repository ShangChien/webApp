```python
from rdkit.Chem import rdMolEnumerator,rdDepictor
from rdkit.Chem import Draw
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.drawOptions.addAtomIndices = True###显示原子索引
IPythonConsole.molSize = 400,400
rdDepictor.SetPreferCoordGen(True)
```

```python
sucrose = "C([C@@H]1[C@H]([C@@H]([C@H]([C@H](O1)O[C@]2([C@H]([C@@H]([C@H](O2)CO)O)O)CO)O)O)O)O"
sucrose_mol = Chem.MolFromSmiles(sucrose)
sucrose_mol
```

```python
primary_alcohol = Chem.MolFromSmarts("C")
sucrose_mol.GetSubstructMatches(primary_alcohol)
sucrose_mol
```
